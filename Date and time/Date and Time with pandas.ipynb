```python
import pandas as pd
```

```python
df = pd.read_csv("ETH_1h.csv")
```

```python
df.head() # observing data first
```

```python
df.shape # numbers of rows and column
```

```python
# randomly locating one value from date column
df.loc[100,'Date']
```

```python
# here the data has two information date and time but they are not in date time format 
# the date and time format code are availabel here
# https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
# based on the information provided on the link 
# we format our date and time
df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d %I-%p')
```

```python
df.head()
```

```python
df.loc[125,'Date'].day_name()# after formatting date and time we used day name function
```

```python
df['Date'].dt.day_name() # gives you the day of all the dates mentioned in data sheet
```

```python
#adding new column to existing data sheet with the name of the day
df['DayOfWeek']=df['Date'].dt.day_name()
```

```python
df
```

```python
# to see earliest and latest dates we use min and max functions
df['Date'].min()
```

```python
df['Date'].max()
```

```python
#time delta gives you the duration between two days
#let's see the time between max and min values obtained above
df['Date'].max() - df['Date'].min()
```

```python
df['Date'].min() - df['Date'].max()
```

```python
# let's filter the data by dates
filt_2020 = (df['Date']>='2020') & (df['Date'] < '2021')
df.loc[filt_2020]
```

```python
filt_2019 = (df['Date']>='2019') & (df['Date']< '2020')
df.loc[filt_2019]
```

```python
#date time column can be set as an index
# to do that we use set_index() function
df.set_index('Date')
```

```python
# if we use inplace = true for the given state above the change will be permanent
# let's go ahead and do it
df.set_index('Date', inplace = True)
```

```python
df
```

```python
# once date is set to be a index 
# we can just pass date to get the datas
df['2019'] # here i have all the dates from 2019
```

```python
# to get the specific data from particular date range we use slice ':'
df['2020-01':'2020-02']
```

```python
# to determine the close value of these dates we just extract close column
df['2020-01':'2020-02']['Close']
```

```python
# to determine the average of these displayed close value we use mean function
df['2020-01':'2020-02']['Close'].mean()
```

```python
# to observe the data on daily base instead of hourly basis 
#first we determine what is value assosiates with daily basis
# lets get high value of each day 
# here we re sample the data by days
high = df['High'].resample('D').max() # D samples in days 'W' samples in week
# https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects
# link provided gives you an ides for the offset of sampling
```

```python
high['2020-01-01']
```

```python
%matplotlib inline
```

```python
high.plot()
```

```python
# resampling by week for min value
low = high = df['High'].resample('W').min()
```

```python
low.plot()
```

```python
# resampling the entire data sets using AGG methods
Weekly_value = df.resample('W').agg({'Close':'mean','High':'max','Low':'min','Volume':'sum'})
```

```python
Weekly_value
```

```python
Weekly_value['Volume'].plot()
```
