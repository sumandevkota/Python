```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeRegressor

```

```python
data = pd.read_csv("/Users/sdevkota01/Downloads/Python/Housing/Me/melb_data.csv")
df = pd.DataFrame(data)
```

```python
features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
X = df[features]
print(X)
y = df.Price
print(y)
```

```python
X.describe()
```

```python
X.head(10)
```

```python
model = DecisionTreeRegressor(random_state=1)
```

```python
model.fit(X,y)
```

```python
solution = model.predict(X)
print(solution)
```

```python
y
```

```python
solution2 = pd.DataFrame(solution)
solution2
```

```python
y2 = pd.DataFrame(y)
y2
```

```python
result = pd.concat([solution2, y2], axis = 1)
result
```

```python
final = pd.concat([X, result], axis = 1)
final
```

```python
final.rename(columns = {0:"Train_Price", "Price":"Predicted_Price"}, inplace = True)
final
```
